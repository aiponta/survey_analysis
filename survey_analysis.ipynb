## 【分析】認知度_前回比較

```python
# q4（認知度）前回比較
import collections

q4_list = []
last_q4_list = []

q4_c = collections.Counter(survey_result['Q4'])
last_q4_c = collections.Counter(last_survey_result['Q4'])

for i in range(4):
    q4_list.append(q4_c[i+1])
    last_q4_list.append(last_q4_c[i+1])

# 表作成
df_q4 = pd.DataFrame(
    data = zip(last_q4_list, q4_list),
    columns=["前回", "今回"],
    index = ['詳しく', '少し', '名前だけ', '知らない']
)
print("表１：認知度比較（前回と今回）")
df_q4.style.background_gradient(cmap='Reds', low=.5, high=2, axis =1)
```

```python
# 地域別q4（認知度）前回
# sc5とq4
last_map_q4_list = [[0 for i in range(4)] for j in range(5)]

for i in range(5):
    for j in range(4):
        last_map_q4_list[i][j] = sum((last_survey_result['Q4'] == j+1) & (last_survey_result['SC5'] == i+1))

# 表作成
df_last_map_q4 = pd.DataFrame(
    data = (last_map_q4_list[0], last_map_q4_list[1], last_map_q4_list[2], last_map_q4_list[3], last_map_q4_list[4]),
    columns = ['詳しく', '少し', '名前だけ', '知らない'],
    index = ['東京23区内', '東京23区外', '横浜', '川崎', '神奈川その他']
)
print("表2：認知度地域別比較（前回）")
df_last_map_q4.style.background_gradient(cmap='Reds', low=0.5, high=2, axis =0)
```

```python
# 地域別q4（認知度）今回
# sc5とq4
map_q4_list = [[0 for i in range(4)] for j in range(5)]

for i in range(5):
    for j in range(4):
        map_q4_list[i][j] = sum((survey_result['Q4'] == j+1) & (survey_result['SC5'] == i+1))

# 表作成
df_map_q4 = pd.DataFrame(
    data = (map_q4_list[0], map_q4_list[1], map_q4_list[2], map_q4_list[3], map_q4_list[4]),
    columns = ['詳しく', '少し', '名前だけ', '知らない'],
    index = ['東京23区内', '東京23区外', '横浜', '川崎', '神奈川その他']
)
print("表2：認知度地域別比較（今回）")
df_map_q4.style.background_gradient(cmap='Reds', low=0.5, high=2, axis =0)
```
